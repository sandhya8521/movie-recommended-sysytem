```python
from statsmodels.tsa.arima.model import ARIMA

# ARIMA model (p, d, q) = (5, 1, 0) based on trial and error for optimal parameters
model = ARIMA(ratings_per_month, order=(5, 1, 0))
model_fit = model.fit()

# Forecasting future ratings (12 months ahead)
forecast = model_fit.forecast(steps=12)

# Plotting the original time series and the forecast
plt.figure(figsize=(10, 6))
ratings_per_month.plot(label='Original Data')
forecast.plot(label='Forecast')
plt.title('ARIMA Forecast - Movie Ratings')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# SARIMA model (p, d, q) x (P, D, Q, s) where s is the seasonality (monthly seasonality = 12)
sarima_model = SARIMAX(ratings_per_month, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_model.fit()

# Forecasting the next 12 months
sarima_forecast = sarima_fit.forecast(steps=12)

# Plotting the original time series and SARIMA forecast
plt.figure(figsize=(10, 6))
ratings_per_month.plot(label='Original Data')
sarima_forecast.plot(label='SARIMA Forecast')
plt.title('SARIMA Forecast - Movie Ratings')
plt.legend()
plt.show()
```

```python
import ast

def parse_genres(x):
    try:
        genres = ast.literal_eval(x)
        return [d['name'] for d in genres]
    except:
        return []

movies_metadata['genres_list'] = movies_metadata['genres'].apply(parse_genres)

# Explode the genres into separate rows
genres_exploded = movies_metadata.explode('genres_list')

# Count of movies per genre
genre_counts = genres_exploded['genres_list'].value_counts()

# Plotting genre distribution
plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
